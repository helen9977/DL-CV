```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

PosNum = 820
NegNum = 1931
winSize = (192,104)
blockSize = (16,16)
blockStride = (8,8)
cellSize = (8,8)
nBin = 9

hog = cv2.HOGDescriptor(winSize,blockSize,blockStride,cellSize,nBin)

svm = cv2.ml.SVM_create()

featureNum = 9936 #int (((192-16)/8+1)*((104-16)/8+1)*4*9)
featureArray = np. zeros(((PosNum+NegNum),featureNum),np.float32)
labelArray = np.zeros(((PosNum+NegNum),1),np.int)

for i in range (0,PosNum):
    fileName = 'pos\\image'+str(i+1)+'.jpg'
    img = cv2.imread(fileName)
    hist = hog.compute(img,(8,8))
    for j in range (0,featureNum):
        featureArray[i,j] = hist[j]
        
    labelArray[i,0] = 1
    
for i in range (0,NegNum):
    fileName = 'neg\\image'+str(i+1)+'.jpg'
    img = cv2.imread(fileName)
    hist = hog.compute(img,(8,8))
    for j in range (0,featureNum):
        featureArray[i+PosNum,j] = hist[j]
        
    labelArray[i,0] = -1
    

svm.setType(cv2.ml.SVM_C_SVC)
svm.setKernel(cv2.ml.SVM_LINEAR)
svm.setC(0.01)


ret = svm.train(featureArray,cv2.ml.ROW_SAMPLE,labelArray)
alpha = np.zeros((1),np.float32)
rho = svm.getDecisionFunction(0,alpha)
print (rho)
print (alpha)
alphaAarry = np.zeros((1,1),np.float32)
supportAarry = np.zeros((1,featureNum),np.float32)
reslutAarry = np.zeros((1,featureNum),np.float32)
alphaAarry[0,0]=alpha
reslutAarry = -1 * alphaAarry*supportAarry

myDetect = np.zeros((9937),np.float32)
for i in range(0,9936):
    myDetect[i] =  reslutAarry[0,i]
myDetect[9936] = rho [0]
myHog = cv2.HOGDescriptor(winSize,blockSize,blockStride,cellSize,nBin)
myHog.setSVMDetector(myDetect)

imageSrc = cv2.imread('Test.jpg',1)
objs = myHog.detectMultiScale(imageSrc,0,(8,8),(32,32),1.05,2)

print (objs)
#x = int(objs[0][0])
#y = int(objs[0][1])
#w = int(objs[0][2])
#h = int(objs[0][3])

#cv2.rectangle(imageSrc,(x,y),(x+w,y+h),(255,0,0),2)
#cv2.imshow('dst',imageSrc)
#cv2.waitKey(0)
```
